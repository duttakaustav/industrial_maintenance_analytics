## 6. Performance Measures

The project evaluates machine performance using a set of maintenance and operational Key Performance Indicators (KPIs).

These measures provide a quantitative view of machine utilization, reliability, and overall operating performance.

The main measures used in this study are:

- Runtime: Total operating time of a machine.
- Downtime: Total period during which a machine is unavailable.
- Availability: Percentage of time that a machine remains available for operation.
- Failure Episodes: Number of recorded machine failure events.
- Mean Time Between Failures (MTBF): Average operating time between failure events.
- Mean Time To Repair (MTTR): Average time required to restore a machine after a failure.
- Energy Consumption: Total electrical energy used during operation.
- Average Temperature: Mean operating temperature during the observation period.
- Average Vibration: Mean vibration level during the observation period.

### Runtime

Runtime represents the total period during which a machine remains operational.

Higher runtime values generally indicate greater machine utilization.

### Downtime

Downtime represents the total period during which a machine is unavailable due to maintenance activities, failures, or operating interruptions.

Higher downtime values generally indicate reduced operating performance.

### Availability

Availability represents the proportion of operating time during which a machine remains available for use.

Availability is directly influenced by downtime.

Higher availability values generally indicate better machine performance.

### Failure Episodes

Failure episodes represent the number of recorded machine failure events during the observation period.

Higher values generally indicate reduced reliability and increased maintenance attention requirements.

### Mean Time Between Failures (MTBF)

MTBF represents the average operating time between failure events.

Higher MTBF values generally indicate improved machine reliability.

### Mean Time To Repair (MTTR)

MTTR represents the average time required to restore a machine after a failure event.

Lower MTTR values generally indicate faster recovery and improved maintenance effectiveness.

### Energy Consumption

Energy consumption represents the amount of electrical energy used during machine operation.

Monitoring energy consumption provides additional insight into machine utilization and operating efficiency.

### Average Temperature

Average temperature represents the mean operating temperature recorded during the observation period.

Elevated temperature values may indicate increased operating stress or reduced cooling effectiveness.

### Average Vibration

Average vibration represents the mean vibration level recorded during the observation period.

Elevated vibration values may indicate mechanical wear, imbalance, or alignment issues.

### Measure Interpretation

These measures provide important context for:

- Condition Review
- Machine Health Evaluation
- Maintenance Guidance
- Engineering Dashboard Results

By examining multiple measures together, the project develops a broader understanding of machine performance and operating condition.

The following code calculates the performance measures and displays a summary of the results.

## 7. Condition Review Method

Condition review is used to identify unusual machine operation before it develops into a more serious operating problem.

The project examines two commonly used machine condition indicators:

- Temperature
- Vibration

### Temperature Review

Machine temperature records are compared against established operating limits.

Elevated temperatures may result from:

- Reduced cooling effectiveness
- Increased mechanical load
- Restricted airflow
- Friction-related heating

Temperature alerts are recorded when operating limits are exceeded and are later used during machine health evaluation.

### Vibration Review

Machine vibration records are compared against established operating limits.

Elevated vibration levels may indicate:

- Bearing wear
- Shaft misalignment
- Mechanical looseness
- Equipment imbalance

Vibration alerts are recorded when operating limits are exceeded and contribute to machine health evaluation.

### Alert Episode Logic

The project records alert episodes rather than individual abnormal measurements.

An alert episode represents a continuous period of unusual machine operation.

This approach helps prevent excessive alert counts that may occur when multiple measurements are recorded during the same operating event.

### Engineering Purpose

Condition review provides an additional view of machine operating behaviour beyond the performance measures discussed earlier.

The resulting alert episodes contribute to machine health evaluation and maintenance guidance.

The following code identifies and summarizes the alert episodes detected during the study.

````python
import pandas as pd
from IPython.display import display

condition_alerts = pd.read_csv(
    "../results/condition_alerts.csv"
)

display(condition_alerts)
````

## 8. Machine Health Evaluation

Machine health evaluation is used to determine the overall operating status of each machine by combining multiple performance and condition indicators.

Rather than relying on a single measure, the evaluation considers operating history, alert activity, and machine condition patterns.

### Evaluation Inputs

The machine health evaluation incorporates:

- Temperature conditions
- Vibration conditions
- Downtime accumulation
- Alert episodes

Together, these indicators provide a broader view of machine condition than any individual measure alone.

### Health Status

Machines are classified into health categories based on the severity of observed operating conditions.

**Healthy**

The machine operates within expected limits and shows no significant operating concerns.

**Warning**

The machine shows unusual operating behaviour that should be monitored and reviewed.

**Critical**

The machine shows severe operating issues and requires immediate attention.

### Risk Level

In addition to health status, a risk level is assigned to each machine.

The risk level helps identify which machines require higher maintenance priority.

Typical classifications include:

- Low Risk
- Medium Risk
- High Risk

### Engineering Purpose

Machine health evaluation converts operating data and alert information into a clear summary of machine condition.

This provides a practical basis for maintenance planning and maintenance guidance.

### Evaluation Results

The evaluation process distinguishes between:

- Machines operating within acceptable limits
- Machines requiring additional monitoring
- Machines requiring immediate attention

The resulting classifications are used to develop maintenance guidance.

The following code evaluates machine health status and displays a summary of the results.

````python
import pandas as pd
from IPython.display import Markdown, display

asset_health = pd.read_csv(
    "../results/asset_health_report.csv"
)

for _, row in asset_health.iterrows():

    display(Markdown(f"### {row['machine_id']}"))

    print(f"Health Status: {row['health_status']}")
    print(f"Risk Level: {row['risk_level']}")
    print(f"Alert Episodes: {row['alert_episodes']}")

    print("\nDetected Issues:")
    print(row["detected_issues"])

    print("\nRecommended Actions:")
    print(row["recommended_actions"])

    print()
````

### Understanding Alert Episodes

Alert Episodes and Failure Episodes represent different types of events.

- Failure Episodes represent machine failures used in reliability evaluation and performance measures.
- Alert Episodes represent abnormal operating conditions identified through condition review.

A single failure event may be associated with multiple alert episodes as machine condition changes over time. Therefore, Alert Episodes may be greater than Failure Episodes, and the two values are not expected to match.
